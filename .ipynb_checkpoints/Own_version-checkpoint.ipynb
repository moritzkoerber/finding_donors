## Imports

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Pretty display for notebooks
%matplotlib inline
```

Load data:

```python
df = pd.read_csv("census.csv")

income_raw = df['income']
features_raw = df.drop('income',axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
skewed = ['capital-gain','capital-loss']
df_log = lambda x: np.log(x+1)
features_log = features_raw
features_log[skewed] = df_log(df[skewed])

plt.hist(features_log['capital-gain'], bins = 20)
plt.ylim(0,2000)
plt.show()
plt.hist(features_log['capital-loss'], bins = 20)
plt.ylim(0,2000)
plt.show()
```

```python
# normalizing
scaler = MinMaxScaler()
nums = features_log._get_numeric_data().columns
features_log_trans = features_log
features_log_trans[nums] = scaler.fit_transform(features_log[nums])
```
